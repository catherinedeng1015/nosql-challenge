### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
hygiene_20_count = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {hygiene_20_count}")
# Display the first document in the results using pprint
results = establishments.find(query)
first_document = results.next() 
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
results_list = list(results) 
hygiene_df = pd.DataFrame(results_list)
# Display the number of rows in the DataFrame
print("Number of rows in the DataFrame:", len(hygiene_df))
# Display the first 10 rows of the DataFrame
print(hygiene_df.head(10))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {"LocalAuthorityName": "City of London Corporation","RatingValue": {"$gte": 4}}

# Use count_documents to display the number of documents in the result
rating_value_count = establishments.count_documents(query)
print(f"Number of establishments in London with a RatingValue >= 4: {rating_value_count}")
# Display the first document in the results using pprint
results = establishments.find(query)
first_document = results.limit(1)
for doc in first_document:
    pprint(doc)
```

```python
# Convert the result to a Pandas DataFrame
results_list = list(establishments.find(query))
df = pd.DataFrame(results_list)

# Display the number of rows in the DataFrame
print(f"Number of establishments found: {len(df)}")

# Display the first 10 rows of the DataFrame
print(df.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.482400  
longitude = -0.011886

query_hygiene = {
    "RatingValue": 5,
    "geocode.latitude": {
        "$gte": str(float(latitude) - 0.1), 
        "$lte": str(float(latitude) + 0.1)
    },
    "geocode.longitude": {
        "$gte": str(float(longitude) - 0.1), 
        "$lte": str(float(longitude) + 0.1)
    }
}
sort = [("scores.Hygiene", 1)] 
results_hygiene = establishments.find(query_hygiene).sort(sort).limit(5)

# Print the results
for data in results_hygiene:
    pprint(data)
```

```python
# Convert result to Pandas DataFrame
results_hygiene_1 = list(establishments.find(query_hygiene).sort("scores.Hygiene", 1).limit(5))
hygiene_df_1 = pd.DataFrame(results_hygiene_1)
print(len(hygiene_df_1))
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1 }}]

# Print the number of documents in the result
results = list(establishments.aggregate(pipeline))
print("Number of Local Authorities with hygiene score of 0:", len(results))
# Print the first 10 results
for i, result in enumerate(results[:10]):
    print(f"{i + 1}. Local Authority: {result['_id']}, Count: {result['count']}")
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print("Number of rows in the DataFrame:", len(df))
# Display the first 10 rows of the DataFrame
print(df.head(10))
```
